# Исследование данных интернет-магазина компьютерных игр «Стримчик»

Интернет-магазин «Стримчик» продаёт по всему миру компьютерные игры. 

Из открытых источников доступны исторические данные до 2016 год о:
- продажах игр,
- оценки пользователей и экспертов,
- жанры,
- платформы (например, `Xbox` или `PlayStation`),
- рейтинг игры по  `ESRB` (Entertainment Software Rating Board)

`ESRB` - ассоциация, определяющая возрастной рейтинг компьютерных игр в США и Канаде. `ESRB` оценивает игровой контент и присваивает ему подходящую возрастную категорию, например, «Для взрослых», «Для детей младшего возраста» или «Для подростков».

**Требуется:**
- выявить определяющие успешность игры закономерности,
- отработать принцип работы с данными для последующих прогнозов.

**Это позволит сделать ставку на потенциально популярный продукт и спланировать рекламные кампании.**

### Предобработка данных

#### Замена названий столбцов

Приведем названия столбцов таблицы к нижнему регистру :

```python
games.columns = games.columns.str.lower()
games.head()
```

#### Замена типов данных

Посмотрим значения `user_score` :


```python
games['user_score'].value_counts()
```

Посмотрим, сколько в нем пропущенных значений:

```python
# let's see the missing values
print('Количество пропусков:', games['user_score'].isna().sum())
```

Заметим, что в `user_score` есть значения `tbd` (“to be determined”), которые тоже означают, что значения еще не определены. Заменим `tbd` на NaN и поменяем тип данных столбца на `float64` :

```python
games['user_score'] = games['user_score'].replace('tbd', np.nan)
# checking the missing values again
print('Теперь пропусков:',games['user_score'].isna().sum())

# changing the data type of the column to float64
games['user_score'] = pd.to_numeric(games['user_score'])
```

Поменяем тип данных для `year_of_release` на `Int64` :

```python
games['year_of_release'] = games['year_of_release'].astype('Int64')  
games.info()
```

#### Обработка пропусков

Посмотрим данные о пропусках :

```python
games.isna().sum()
```

Удалим строки, где пропуски в `name`, `genre`, `year_of_release` (так как их довольно мало от общего числа строк):

```python
games = games.dropna(subset=['name', 'genre', 'year_of_release']).reset_index(drop=True)
```

Проверим :

```python
games.isna().sum()
```

Так как рейтинг ESRB присваивается только для игр в Америке и Канаде, то заменим остальные пропуски на `no_rating` для игр в остальных регионах :

```python
games['rating'] = games['rating'].fillna('no_rating')
```

Проверим:

```python
games.isna().sum()
```

#### Обработка дубликатов

Посмотрим явные дубликаты :

```python
print('Количество явных дубликатов:', games.duplicated().sum())
```

Будем искать скрытые дубликаты. Будем искать строчки, в которых название игры, платформа и год выпуска - одинаковые. 

```python
games[games[['name', 'platform','year_of_release']].duplicated(keep=False)]
```

Удалим вторую строку игры `Madden NFL 13` и проверим, что дубликатов после удаления действительно больше нет:

```python
games = games.drop(index=15969)
games[games[['name', 'platform','year_of_release']].duplicated(keep=False)]
```

Будем искать скрытые дубликаты по нечисловым столбцам. Посмотрим уникальные значения и проверим данные на опечатки :

```python
games['platform'].unique()
```

```python
games['genre'].unique()
```

```python
games['rating'].value_counts()
```

Заметим, что есть значения рейтинга `K-A (kids - adults)` , которые на самом деле скрытые дубликаты `E (for everyone)`. Заменим их на `E` :

```python
games.loc[games['rating'] == "K-A", 'rating'] = 'E'
```

Проверим:

```python
games['rating'].value_counts()
```

Посмотрим столбцы с оценками пользователей и критиков, чтобы не было выбивающихся значений (выше возможной оценки):

```python
games['critic_score'].value_counts()
```

```python
games['user_score'].value_counts()
```

Здесь все в порядке. 

#### Суммарные продажи во всех регионах 

Создадим столбец `all_sales`, где будем сумма продаж по всем регионам :

```python
games['all_sales'] = games['na_sales'] + games['eu_sales'] + games['jp_sales'] + games['other_sales']
```

Проверим:

```python
games.head()
```

#### **Общий вывод раздела :**
- названия столбцов приведены к стандартному виду,
- произведены замены типов данных в `user_score` и `year_of_release`,
- обработаны пропуски: удалены строки, где присутвовали пропуски в столбцах `name`, `genre`,`year_of_release`, так как в этих столбцах малый процент пропусков. Строки, где пропуски в `user_score` и `critic_score` трогать не будем, так как в этих столбцах около 50% пропусков от всех значений. Пропуски в `rating` заполнены на `no_rating`,
- были обнаружены скрытые дубликаты в информации об оценках пользователей (`user_score`). После чего произведена замена значений `tbd` на пропуски, обработаны остальные дубликаты,
- создан столбец `all_sales` с суммарной информацией о продажах по всем регионам

#### Продажи по платформам в динамике

Посмотрим распределение по количеству продаж на разных платформах :

```python
games.pivot_table(
    index='platform',
    values='all_sales',
    aggfunc='sum'
).sort_values(
    by='all_sales',
    ascending=False
).plot(
    kind='bar',
    ylabel='общие продажи',
    xlabel='платформы',
    title='Продажи по платформам',
    rot=0,
    stacked=True,
    width=0.8,
    color=colors[0],
    legend=False,
    figsize=(14,8)
);
plt.xticks(rotation=45);
```

Выберем платформы с наибольшими суммарными продажами (PS2, X360, PS3, Wii) и построим распределение по годам :

```python
platform_list = ["PS2", "X360", "PS3", "Wii"]

for plat in platform_list:
    games.query(
        f'platform == "{plat}"'
    ).pivot_table(
        index='year_of_release',
        values='all_sales',
        aggfunc='sum'
    ).plot(
        kind='bar',
        title= f'Продажи на {plat}',
        ylabel='общие продажи',
        xlabel='год',
        rot=0,
        color=colors[0],
        stacked=True,
        width=0.8,
        legend=False,
        figsize=(8,5)
    );    
```

Найдем медианное значение для количества лет "жизни" платформ :

```python
platform_life = games.pivot_table(
    index='platform',
    values='year_of_release',
    aggfunc=['min', 'max']
)
platform_life['life'] = platform_life['max'] - platform_life['min'] + 1
print('Медианное время существования платформы:', platform_life['life'].median())
```

**Вывод:**
- платформы "живут" около 7 лет.

```python
games.query(
    'year_of_release >= 2013'
).pivot_table(
    columns='platform',
    index='year_of_release',
    values='all_sales',
    aggfunc='sum'
).plot(
    kind='bar',
    stacked=True,
    color=colors,
    figsize=(10,6)
)
plt.title('Распределение продаж по годам с 2013 года')
plt.xlabel('Год')
plt.ylabel('Продажи')
plt.xticks(rotation=0)
plt.show();
```

#### Корреляция между отзывами пользователей и критиков и продажами

### Проверка гипотез

#### Средние пользовательские рейтинги платформ Xbox One и PC одинаковые

Проверим гипотезу о равенстве пользовательских рейтингов платформ Xbox One и PC.

Пользовательские рейтинги платформ Xbox One и PC являются незавсисмыми друг от друга генеральными совокупностями.
Когда сравнивают средние двух генеральных совокупностей, нулевую гипотезу можно сформулировать так: «средние значения совокупностей равны».
А альтернативную - что не равны.

Воспользуемся т-тестом: `scipy.stats.ttest_ind(array1, array2, equal_var, alternative)`.

t-тест работает для выборки любого размера

**Нулевая гипотеза** - средние пользовательские рейтинги платформ Xbox One и PC одинаковые

**Альтернативная гипотеза** - средние пользовательские рейтинги платформ Xbox One и PC разные

Проведем двусторонний `t-test` с уровнем значимости `alpha = 0.01`

```python
alpha = 0.01 # уровень статистической значимости

results = st.ttest_ind( 
    games_new.loc[games['platform'] == "XOne"]['user_score'],
    games_new.loc[games['platform'] == "PC"]['user_score'],
    equal_var=False,
    nan_policy='omit'
)

print(f'p-value : {round(results.pvalue, 2)}')

# проверяем p-value
if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Нет оснований отвергнуть нулевую гипотезу')
```

```python
print('Средний рейтинг XOne :',
      round(games_new.loc[games['platform'] == "XOne"]['user_score'].mean(),2)
     )
print('Средний рейтинг PC :',
      round(games_new.loc[games['platform'] == "PC"]['user_score'].mean(),2)
     )
print(
    'Различие в средних:',
    round((games_new.loc[games['platform'] == "XOne"]['user_score'].mean() - 
 games_new.loc[games['platform'] == "PC"]['user_score'].mean()) /games_new.loc[games['platform'] == "XOne"]['user_score'].mean()*100,2),
    '%'
)
```

Средние пользовательские рейтинги платформ Xbox One и PC отличаются на 4%, что показывает, что данные не противоречат нулевой гипотезе

**Вывод:**
- На уровне стат.значимости в 1% не удалось отвергнуть нулевую гипотезу о равенстве средних пользовательских рейтингов платформ Xbox One и PC.

#### Средние пользовательские рейтинги жанров `Action` и ` Sports `


Проверим гипотезу о равенстве средних пользовательских рейтингоа жанров `Action` и ` Sports `.

**Нулевая гипотеза** - средние пользовательские рейтинги жанров `Action` и ` Sports ` одинаковые

**Альтернативная гипотеза** - средние пользовательские рейтинги жанров `Action` и ` Sports ` разные

Проведем `t-test` с уровнем значимости `alpha = 0.01`

```python
alpha = 0.01 # уровень статистической значимости

results = st.ttest_ind( 
    games_new.loc[games['genre'] == "Sports"]['user_score'],
    games_new.loc[games['genre'] == "Action"]['user_score'],
    equal_var=False,
    nan_policy='omit'
)

print(f'p-value: {round(results.pvalue, 20)}')

# проверяем p-value
if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Нет оснований отвергнуть нулевую гипотезу')
```

Посмотрим средние значения пользовательских рейтингов жанров `Action` и `Sports` :

```python
print('Средний рейтинг Sports :', round(games_new.loc[games['genre'] == "Sports"]['user_score'].mean(),2))
print('Средний рейтинг Action :', round(games_new.loc[games['genre'] == "Action"]['user_score'].mean(), 2))
print('Различие в средних рейтингах:', round((games_new.loc[games['genre'] == "Action"]['user_score'].mean() - games_new.loc[games['genre'] == "Sports"]['user_score'].mean())/
 games_new.loc[games['genre'] == "Action"]['user_score'].mean()*100, 2),'%')
```

Средние рейтинги этих двух жанров отличаются на 23%, что действительно противоречит нулевой гипотезе о равенстве средних

**Вывод:**
-  На имеющихся данных, на уровне значимости 1% есть основания отвергнуть нулевую гипотезу о равенстве средних значений пользовательских рейтингов (отзывов) жанров `Action` и `Sports` в пользу альтернативной гипотезы.
- Средний рейтинг жанра `Action` больше `Sports` на 23%.
