Will add new column `year` which will be useful for further analysis.

For every client, we will introduce a new column named `treatment_number`, which will indicate the count of treatments.

- **A treatment** will encompass **all the services** rendered to a client within a timeframe **where the gap between any two consecutive dates is no greater than 14 days**. 
- If there are two visits for a client with a gap exceeding 14 days between them, we will consider them as separate treatments.

```python
# Sort the DataFrame by client id ('insured') and visit date ('service_date') in ascending order
med_data = med_data.sort_values(by=['insured', 'service_date'])

# Calculate the difference in days between visits for each client
med_data['days_between_visits'] = med_data.groupby('insured')['service_date'].diff().dt.days

# Initialize the treatment number as 1
treatment_number = 1

# Create an empty list to store treatment numbers for each client
treatment_numbers = []
current_client = None

# Iterate through the DataFrame and assign treatment numbers
for index, row in med_data.iterrows():
    if current_client is None or current_client != row['insured']:
        current_client = row['insured']
        treatment_number = 1
    else:
        if row['days_between_visits'] > 14:
            treatment_number += 1
    treatment_numbers.append(treatment_number)

# Add the treatment_numbers list as a new column in the DataFrame
med_data['treatment_number'] = treatment_numbers


med_data.head(20) # checking
```

#### **Section's conclusion:**
- uploaded dataframe has some missing values and no duplicates,
- the column's `service_date` data type is changed to datetime,
- not all dates are indicated for 2022 (only from January to September),
- column `year` is created for further analysis,
- the treatment was defined as :
    - **all the services** rendered to a client within a timeframe **where the gap between any two consecutive dates is no greater than 14 days**. 
    - if there are two visits for a client with a gap exceeding 14 days between them, we will consider them as separate treatments.
- for every client the column named `treatment_number`(which indicates the count of treatments) is created. All rows for each client with close dates have the same treatment_number.

### Treatment cost 

#### Definition of treatment cost

**In this work, we will decide that the treatment** will encompass **all the services** rendered to a client within a timeframe **where the gap between any two consecutive dates is no greater than 14 days**. 

- If there are two visits for a client with a gap exceeding 14 days between them, we will consider them as separate treatments.

Thus, **the cost of treatment** for this client will be equal to **the sum of all costs for services provided within a timeframe where the gap between any two consecutive dates is no greater than 14 days**.

- **We will also only look at the relevant months from 2021 to 2022 (from January to September)**

#### Treatment cost distributions

Let's plot the dustribution of treatment costs in 2021 and 2022:

```python
# calculating all treatments' costs for each client in each year
treatment_cost = med_data.query('service_date < "2021-09-01" or service_date >= "2022-01-01"').groupby(['insured', 'treatment_number', 'year']).agg({'service_amount':'sum'}).reset_index()

# plotting the distributions of treatments' costs in each year

sns.histplot(
    treatment_cost.query('service_amount <= 60000').loc[treatment_cost['year']==2021]['service_amount'], 
    stat="density",
    common_norm=False,
    kde=True,
    bins=range(1, 60000, 1000)
)
sns.histplot(
    treatment_cost.query('service_amount <= 60000').loc[treatment_cost['year']==2022]['service_amount'],
    color='green',
    alpha=0.3,
    stat="density",
    common_norm=False,
    kde=True,
    bins=range(1, 60000, 1000)
)
plt.legend(labels=['2021', '2022'])
plt.ylabel('density')
plt.xlabel('treatment cost')
plt.title('Distribution of treatment costs in 2021 and 2022');
```

Let's see boxplots for each treatment_number. 

```python
# looking at treatment costs for each treatment_number
sns.boxplot(data=treatment_cost, x='treatment_number', y='service_amount', hue='year', palette=colors)
plt.xlabel('treatment number')
plt.ylabel('treatment cost')
plt.title('Treatment costs')
plt.ylim(0, 30000);
```

#### Hypothesis testing : average treatment cost in 2021 and 2022

Do the treatment cost distributions have equal means or not? Let's make a decision using hypothesis testing.

As we can see, these distributions are not normal and have many outliers. Thus, we will use the Mann-Whitney test to test the hypothesis.

- **H0 – the mean costs of treatment in 2021 and 2022 are equal.**
- **H1 - the mean costs of treatment in 2021 and 2022 are not equal.**

Alpha - 0.05

We will use a one-sided test, assuming that the cost of treatment in 2021 was less than in 2022.

```python
# testing the hyposhesis about mean treatment cost in 2021 and 2022

stats.ttest_ind(
    treatment_cost.loc[treatment_cost['year']==2021]['service_amount'],
    treatment_cost.loc[treatment_cost['year']==2022]['service_amount'],
    alternative='less'
)
```

The p-value is much less than 0.05, so we can reject the null hypothesis that average treatment costs are equal in 2021 and 2022.

Let's see how median trestment cost increased in 2022:

```python
(
    treatment_cost.loc[
        treatment_cost['year']==2022
    ]['service_amount'].median() - treatment_cost.loc[
        treatment_cost['year']==2021
    ]['service_amount'].median()
)/ treatment_cost.loc[
    treatment_cost['year']==2021
]['service_amount'].median()*100
```

#### **Section's conclusion:**
- the treatment cost term is defined,
- the distributions of treatment costs tested using Mann-Whitney test (one-sided):
    - the p-value is much less than 0.05, so we can reject the null hypothesis that average treatment costs are equal in 2021 and 2022.
    - **average treatment cost is higher in 2022 than in 2021,** 
    - median treatment cost increased in 2022 by 11%.

##### **Hypothesis testing: average cost of services in 2021 and 2022**

Now will look at distributions of all services' costs:

```python
# all services' costs in 2021:
costs_2021 = med_data.query('service_date < "2021-09-01" and service_number==1').groupby(['service_name_corrected']).agg({'service_amount':'median'}).reset_index()
# all services' costs in 2022:
costs_2022 = med_data.query('service_date >= "2022-01-01" and service_number==1').groupby('service_name_corrected').agg({'service_amount':'median'}).reset_index()

# plotting normalised histograms
sns.histplot(
    costs_2021['service_amount'],
    stat="density",
    common_norm=False,
    kde=True,
    bins=range(0, 10000, 200)
)
sns.histplot(
    costs_2022['service_amount'],
    color='green',
    alpha=0.5,
    stat="density",
    common_norm=False,
    kde=True,
    bins=range(0, 10000, 200)
)
plt.xlim(0, 10000)
plt.legend(labels=['2021', '2022'])
plt.xlabel('service cost')
plt.ylabel('density')
plt.title('Distribution of the cost of all services');
```

Do the service cost distributions have equal means or not? Let's make a decision using hypothesis testing.

As we can see, these distributions are not normal and have many outliers. Thus, we will use the Mann-Whitney test to test the hypothesis.

Let's see if these two distributions have equal average service costs.
We will use a one-sided test, assuming that the average cost of services in 2021 was less than in 2022.

- **H0 - The average cost of services in 2021 is the same as in 2022.**
- **H1 - The average cost of services in 2021 is not equal to 2022.**

Alpha = 0.05

```python
# carrying out U test
stats.mannwhitneyu(
    costs_2021['service_amount'],
    costs_2022['service_amount'],
    alternative='less'
)
```

p-value is less than 0.05, null hypothesis is declined.

Let's find the percentage difference in the median cost of services in 2022 compared to 2021:

```python
(costs_2022['service_amount'].median() - costs_2021['service_amount'].median())/costs_2021['service_amount'].median()*100
```

The median cost of services in 2022 increased by 7.7%.

##### **Subsection's conclusion:**
- p-value is less than 0.05, null hypothesis is declined, 
- the average cost of services in 2021 is less than in 2022,
- the median cost of services in 2022 increased by 7.7%.

- will construct the distribution of the percentage of expensive services per treatment (in 2021 and 2022),

```python
# estimating the boundary for expensive services in 2021
med_data.query(
    'service_date >= "2022-01-01"'
)['service_amount'].quantile(q=0.75)
```

**Will define**
- expensive services for **2021** - with the price **more than 1470**
- expensive services for **2022** - with the price **more than 1540**

##### Calculating  for each treatment the percentage of expensive services in it

Will create columns `expensive_or_not_2021` and `expensive_or_not_2022` with `True/False` for determing if the service is expensive or not for each year. 

```python
med_data['expensive_or_not_2021'] = ((med_data['service_amount']/med_data['service_number']) > 1470) & (med_data['year']==2021)
med_data['expensive_or_not_2022'] = ((med_data['service_amount']/med_data['service_number']) > 1540) & (med_data['year']==2022)
med_data.head()
```
